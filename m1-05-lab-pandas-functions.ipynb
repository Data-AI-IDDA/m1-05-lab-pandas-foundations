<h1>Task 1</h1>

```python
import pandas as pd
import random

attendance_raw = []

cohorts = ["alpha", "beta", "gamma"]
numbers = list(range(7))
weights = [1, 1, 1, 1, 1, 1, 10]


for i in range(1, 25): 
    record = {
        "student_id": f"S{i:03}",
        "cohort": random.choice(cohorts),
        "attended_sessions": random.choices(numbers, weights=weights, k=1)[0],
        "expected_sessions": 6
    }
    
    attendance_raw.append(record)

attendance = pd.DataFrame(attendance_raw)

print(attendance.head())
print()
attendance.describe()
```

<h1>Task 2</h1>

```python
attendance_indexed = attendance.set_index('student_id')
print(attendance_indexed.head(5))
```

```python
std_ids = random.sample([f'S{i:03}' for i in range(1,51)],15)
excused_absences = pd.Series({i : random.randint(1,3) for i in std_ids})
excused_absences
```

```python
attendance_indexed['adjusted_attendance'] = attendance_indexed['attended_sessions'] + excused_absences
attendance_indexed
```

```python
attendance_indexed['adjusted_attendance'] = attendance_indexed['adjusted_attendance'].fillna(
    attendance_indexed['attended_sessions'])
attendance_indexed
```

<h1>Task 3</h1>

```python
import random

random_ids = random.sample([f'S{i:03}' for i in range(1,25)], 5)

for student_id in random_ids:
    attendance_indexed.loc[student_id, 'cohort'] = random.choice(
        [' Beta ', 'ALPHA', ' alpha', 'gAmMa', 'beTa', 'gamma ']
    )
attendance_indexed['cohort'].unique()
```

```python
attendance_indexed['cohort'] = attendance_indexed['cohort'].str.strip().str.lower()
attendance_indexed['cohort'].unique()
```

<h1>Task 4</h1>

```python
low_attendance = attendance_indexed.where(attendance_indexed['attended_sessions'] < attendance_indexed['expected_sessions']).dropna()
low_attendance
```

```python
attendance_indexed.groupby('cohort')['attended_sessions'].mean()
```
