```python
import pandas as pd
import numpy as np
import random
```

```python
attendance_raw = []
cohort =  ["alpha", "beta", "gamma"]
for i in range(1, 25):
    records = {'student_id':f"S{i:03d}", 'cohort': np.random.choice(cohort), 'attended_sessions':np.random.randint(0, 6), 'expected_sessions': 6} 
    attendance_raw.append(records)
```

# Task 4
Filter the DataFrame to students where attended_sessions is below expected_sessions. Store the result in low_attendance. Compute the average attended_sessions by cohort using groupby. Print the summary and verify that cohorts in the summary match the cleaned cohorts.

```python
low_attendance = attendance_indexed.loc[attendance_indexed['attended_sessions'] < attendance_indexed['expected_sessions']]
```

```python
low_attendance
```

```python
avg_att = attendance_indexed.groupby(['cohort'])['attended_sessions'].mean()
```

```python
print(avg_att)
```

```python
summary_cohorts = set(avg_att.index)
data_cohorts = set(attendance_indexed['cohort'].unique())

print("\nValidation:")
print("Cohorts match:", summary_cohorts == data_cohorts)
```
